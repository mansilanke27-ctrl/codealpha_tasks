```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("economicsdataseteda.csv")
df.head()
```

```python
df.columns

```

```python
df = df.dropna()
```

```python
top10 = df.sort_values(by="market_cap_usd", ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x="symbol", y="market_cap_usd", data=top10)
plt.title("Top 10 Crypto by Market Cap")
plt.show()
```

```python
sns.barplot(x="symbol", y="price_usd", data=top10)
plt.title("Top Crypto Prices")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

**Conclusion**

The analysis of the Top 50 cryptocurrencies shows that the market is dominated by a few major coins with very high market capitalization. Price, market cap, and trading volume vary significantly among cryptocurrencies, indicating different levels of market demand and investor interest. Data visualization helped in clearly understanding trends, comparisons, and relationships within the crypto market. Overall, this project demonstrates how visual analytics can convert raw data into meaningful insights for better decision-making.
